- prompt which continuously ask to roll the dice in Y or N format
- If Y then roll the dice and give 2 numbers (n1, n2), then prompt again the question
- If N then quit the game

```python
import random
print('========= Welcome to Dice rolling game =========')

# d1 = [n for n in range(1,7)]
# d2 = [n for n in range(1,7)]

# print(random.choice(d1))
# print(random.choice(range(1,7)))

play = True

while play:
    ask = input("Do you want to roll the dice. Y | N : \n").lower()
    if ask == "y":
        dice = (random.choice(range(1,7)), random.choice(range(1,7)))
        print(dice)
    elif ask == 'n':
        play = False
        print("Thank YOU ! for playing.")
    else :
        print("Invalid choice, try again with 'Y' or 'N'.")
        
                                      
```
