```python
# computer will generate a random number between 1 and 100.
# loop
    # Guess the number the only hint you will get is too high  or too low... this will be in loop untill you find the right number
    #  once the user hits the right number the loop gets over and prompts a congratulations message.

import random

number = random.randint(1,100)
game = True

while game:
    guess = input("Enter the number between 1 and 100 : \n")

    if not guess.isnumeric():
        print("Invalid output")
    elif int(guess) == number:
        game = False
        print("Congratulations...! {} is the correct number".format(guess))
    elif int(guess) <= number:
        print("Too low.")
    elif int(guess) >= number:
        print("Too high.")
```
