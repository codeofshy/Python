```python
class Car:
    def __init__(self, brand, name):
        self.brand = brand
        self.name = name

    def car_info(self):
        print(f'{self.brand} - {self.name}')

car1 = Car('Tata','safari')

        
```

```python
car1.name
```

```python
car1.brand
```

```python
car1.car_info()
```

```python
# encapsulation : getter and setter

class Bank:
    def __init__(self, fund):
        self.fund = fund
        
    # getter
    def get_balance(self):
        print(f'Available balance : {self.fund}')

    #setter
    def add_amount(self, deposit):
        self.fund += deposit
        print(f'Your updated balance is {self.fund}')
```

```python
shayan = Bank(1000)
```

```python
shayan.get_balance()
```

```python
shayan.add_amount(50)
```

```python
shayan.get_balance()
```

```python
# inheritance
class Animal:
    def sound(self):
        print('Animal making sound...!')

class Dog(Animal):
    def details(self):
        self.sound()
        print('This is dog...!')
```

```python
d = Dog()
```

```python
d.details()
```

```python
# multiple inheritance
class A:
    def method_A():
        print("method A")

class B:
    def method_B():
        print("method B")

class C(A,B):
    pass


    
    
```

```python
c1 = C
```

```python
c1.method_A()
```

```python
c1.method_B()
```
